## 데이터 체크하기

```python
%%sql
-- CTAS
DROP TABLE IF EXISTS table_name ;
CREATE TABLE table_name AS
SELECT * FROM ~;
```

```python
%%sql
-- 중복 체크
SELECT COUNT(1)
FROM sale_data;

SELECT COUNT(1)
FROM (
    SELECT DISTINCT *
    FROM sale_data
)a;
```

```python
%%sql
-- CTE : WITH로 임시 테이블을 만드는 방법
WITH ds AS (
    SELECT DISTINCT *
    FROM sale_data
)
SELECT COUNT(1)
FROM ds;
```

```python
%%sql
-- NULL COUNT
SELECT
    COUNT(CASE WHEN order_id is NULL THEN 1 END) order_id_null_count,
    COUNT(CASE WHEN date is NULL THEN 1 END) date_null_count,
    COUNT(CASE WHEN itemcode is NULL THEN 1 END) itemcode_null_count,
    COUNT(CASE WHEN category is NULL THEN 1 END) category_null_count,
    COUNT(CASE WHEN price is NULL THEN 1 END) price_null_count,
    COUNT(CASE WHEN amount is NULL THEN 1 END) amount_null_count
FROM
    sale_data;
```

```python
%%sql

DROP TABLE IF EXISTS revenue_per_month_channel;
CREATE TABLE revenue_per_month_channel
AS
SELECT 
LEFT(ts,7) AS month,
b.channel,
COUNT(DISTINCT b.userid) AS uniqueUsers,
COUNT(DISTINCT CASE WHEN c.refunded is NOT NULL THEN b.userid END) AS paidUsers,
ROUND(paidUsers*100.0/NULLIF(uniqueUsers,0),2) AS conversionRate,
SUM(c.amount) AS grossRevenue,
SUM(CASE WHEN c.refunded is FALSE THEN c.amount ELSE 0 END) AS netRevenue
FROM
    session_timestamp a JOIN user_session_channel b ON a.sessionid=b.sessionid
                        LEFT JOIN session_tansaction c ON a.sessionid=c.sessionid
GROUP BY 1,2
ORDER BY 1,2;
```

```python
SELECT userid , FIRST_VALUE(channel) OVER(order by ts) as first_channel, LAST_VALUE(channel) OVER(order by ts) as last_channel
FROM session_timestamp a JOIN user_session_channel b ON a.sessionid=b.sessionid
group by 1
order by 1;

SELECT 
userid ,
ROW_NUMBER() OVER(partition by userid order by ts) as first_channel,
ROW_NUMBER() OVER(partition by userid order by ts DESC) as last_channel
FROM session_timestamp a JOIN user_session_channel b ON a.sessionid=b.sessionid
WHERE 
ROW_NUMBER() OVER(partition by userid order by ts) = 1 or 
ROW_NUMBER() OVER(partition by userid order by ts DESC) = 1
order by 1;
```

```python
%%sql
-- 날짜별 처음,마지막 판매 물품 first_value,last_value 써서
SELECT 
DISTINCT
date, 
first_value(itemcode) over(partition by date),
last_value(itemcode) over(partition by date)
FROM (SELECT * FROM sale_data order by 1)a
order by 1
LIMIT 10;
```

```python
%%sql
-- 날짜별 처음,마지막 판매 물품 first_value,last_value 써서
SELECT 
DISTINCT
date, 
first_value(itemcode) over(partition by date order by order_id 
                           rows between unbounded preceding and unbounded following),
last_value(itemcode) over(partition by date order by order_id
                         rows between unbounded preceding and unbounded following)
FROM sale_data
order by 1
LIMIT 10;
```

```python
%%sql
-- 날짜별 처음,마지막 판매 물품 ROW_NUMBER 써서
SELECT a.date, a.itemcode as first_value, b.itemcode as last_value
FROM
(SELECT date, itemcode, ROW_NUMBER() OVER(partition by date order by order_id) as seq FROM sale_data) a JOIN
(SELECT date, itemcode, ROW_NUMBER() OVER(partition by date order by order_id DESC) as seq FROM sale_data) b
ON a.date=b.date and b.seq=1;
WHERE a.seq=1; -- ON에 넣어도 되는데 FROM table의 조건은 WHERE에 하는 것이 일반적
```

```python
%%sql
-- 가장 잘팔린 itemcode 상위 10개
SELECT itemcode, ROUND(SUM(amount)/10000.0,2) as Gross_revenue
FROM
sale_data
group by 1
order by 2 DESC
LIMIT 10;
```

```python
%%sql
-- 가장 잘팔린 itemcode 상위 10개
SELECT DISTINCT itemcode, ROUND(SUM(amount) OVER(partition by itemcode)/10000.0,2) as Gross_revenue
FROM
sale_data
order by 2 DESC
LIMIT 10;
```

```python
SELECT a.mon, ROUND((a.promoter-a.passive)*100.0/NULLIF(a.total_count,0),2) as NPS
FROM
(SELECT 
LEFT(ts,7) AS mon, 
COUNT(CASE WHEN score BETWEEN 9 AND 10 THEN 1 END) AS promoter,
COUNT(CASE WHEN score BETWEEN 7 AND 8 THEN 1 END) AS passive,
COUNT(CASE WHEN score BETWEEN 0 AND 6 THEN 1 END) AS detractor,
COUNT(1) AS total_count
FROM ~
group by 1)a;
```
