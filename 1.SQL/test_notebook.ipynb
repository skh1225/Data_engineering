```python
%%sql
SELECT
category,
STRING_AGG(RIGHT(itemcode::varchar,2), ORDER BY price) itemcodes
FROM item_data
group by 1
LIMIT 10;
```

```python
%%sql
SELECT
DISTINCT
category,
STRING_AGG(RiGHT(itemcode::varchar,2),' ') OVER(partition by category order by price rows between unbounded preceding and unbounded following) itemcodes
FROM item_data
LIMIT 10;
```
